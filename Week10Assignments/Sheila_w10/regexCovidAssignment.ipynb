```python
# Load the re module

# Code here...
import re
```

```python
# Create a file object to read from the covid data file
# This file contains cumulative case counts and deaths from 
# covid-19 for every country in the US. It was current as of
# March 30th.

# Code here...
# creating the file object from the covid .txt
fileOne=open("covidData_3.30.20.txt", "r")

# opening the object file to see what is inside
if fileOne.mode=="r":
    content=fileOne.read()
    print(content)
```

```python
# Find all parishes from Louisiana in this file.
# You should make a list that contains only the parish
# names. This list can be created with a single command.
# Be careful to include ALL the parishes in your search.

# Code here...

#searching for all words before ",Louisiana" and storing the list in the variable parishes
parishes=re.findall(r"([\w.\s*]+),Louisiana", content)

#checking the list
#print(parishes)

# creating a list to store the name of all parishes in Lousiana  
parishesList=[]
for parish in parishes:
    
# adding the names to the list without being duplicated
    if parish not in parishesList:
        parishesList.append(parish)   
```

```python
# Close the covid data file object

# Code here...
fileOne.close()
```

```python
# Print out the Louisiana search results

# Code here...
#checking the final list
print(parishesList) 
```

```python
# Create a dictionary that has the Louisiana parish 
# names as keys and the number of times they appear 
# in the data file as values.

# Hint: You can test if a dictionary does or does not
# contain a certain key using:
# myKey in myDict --or-- myKey not in myDict

# Code here...

#creating the directory to store the names and frequencies
parishesDic={}

#counting the number of times each LA parish appears in the file
for parish in parishes:
    if parish in parishesDic:
        parishesDic[parish]+=1
    else:
        parishesDic[parish]=1
        
#printing the directory
print(parishesDic)
```

```python
# Using your dictionary, figure out which of these parishes has
# the most entries in the file. The one with the most entries
# had the earliest observed covid case.
# Jefferson, Orleans, East Baton Rouge, Plaquemines

# Code here...

#getting the maximun value from the directory
mostEn = max(parishesDic,key=parishesDic.get)
print("The LA parish with the most entries in the file is", mostEn)
```

```python
# Figure out how many parishes are in the dictionary. How does 
# this compare to the total number of parishes in the state?

# Code here...
#getting the total number of LA parishes in the directory 
number=len(parishesDic)
print("There are", number, "parishes affected out of the 64 total parishes in LA")
```

```python
# Now, reopen the data file and use a regex search to find the first instance 
# where a Louisiana parish recorded a covid-19 death. This will be the first 
# line in the file that includes Louisiana and has a death count >0. After 
# your search, print the date and the parish. Be sure to also close the input file.
# Example output: ZZZZ parish reported the first covid-19 death in Louisiana on 2020-ZZ-ZZ

# Code here...
fileOne=open("covidData_3.30.20.txt", "r")

if fileOne.mode=="r":
    content=fileOne.read()

#searching for all components on the first line where the word Louisiana and the number 1 (at the end of the line, corresponding to the numbers of deaths) appear 
#at the same time, searching for each component as an independent group
caseOne=re.search(r"([\d.-]+),(\w+),Louisiana,(\d+),(\d+),(1)", content) 

#checking each number group to assign a variable to the ones needed: date and parish
#caseOne.group()

date=caseOne.group(1)
parish=caseOne.group(2)

#printing the final outcome
print(parish,"parish reported the first covid-19 death in Louisiana on", date)
fileOne.close()
```

```python
# Now we need to create a new file that only has records from Louisiana. 
# Extract the lines that pertain to Louisiana and write them to a new file.

# Code here...
fileOne=open("covidData_3.30.20.txt", "r")

#getting the header to include it in the new text file
header=re.search(r"date,(\w.+),deaths", content)
#header.group()
headerTwo=header.group()

headerStr=str(headerTwo)
#print(headerStr)

#creating the new text file
newFile=open("covidLA.txt", "a+")

#copying the header to the new file
newFile.write(headerTwo + "\n")

#searching for all entries/lines with the word Louisiana in it
for x in fileOne:
    entry=re.search(r"([\d.-]+),([\w.\s*]+),Louisiana,(\d+),(\d+),(\d+)", x)
    if entry:
#copying the result in the new file
        newFile.write(entry.group() + "\n")

newFile.close()

newTwo=open("covidLA.txt", "r")
if newTwo.mode=="r":
    cont=newTwo.read()
print(cont)
```

```python
# Lastly, let's imagine that we want to do further downstream analyses with
# the Louisiana data, but our downstream analyses require some formatting chanages.
# Read in the lines from your newly created file with only Louisiana records, reformat them, 
# and create a separate file to hold the newly formatted information. Specifically, the 
# dates in the file should be reformatted from looking like this (2020-MM-DD) to looking 
# like this (MM.DD.2020). Also, the fips codes (in the 4th column) should be removed, and
# Louisiana should be abbreviated as LA. So, this file should end up with entries that look
# like this: 03.25.2020,Morehouse,LA,1,0

# Code here...

#opening the new final with info from LA only
newTwo=open("covidLA.txt", "r")
if newTwo.mode=="r":
    cont=newTwo.read()

#creating the last file to store the new formatted info from LA 
newLastFile=open("newcovidLA.txt", "a+")

#searching for all lines including the word Louisiana 
format=re.findall(r".+-(.+)-(.+),(.+),Louisiana,.*,(.+),(.+)", cont)

#getting each result as a different group/variable
for x in format:
    month=x[0]
    day=x[1]
    parish=x[2]
    cases=x[3]
    deaths=x[4]
#copying each new line in the last file with the new format
    newLastFile.write("%s.%s.2020,%s,LA,%s,%s" %(month,day,parish,cases,deaths)+"\n")

#closing both files
newLastFile.close()
newTwo.close()

#checking the final result
last=open("newcovidLA.txt", "r")
if last.mode=="r":
    conten=last.read()
print(conten)
```
