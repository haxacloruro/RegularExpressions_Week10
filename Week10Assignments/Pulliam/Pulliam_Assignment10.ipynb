```python
# Load the re module
import re
```

```python
# Create a file object to read from the covid data file
# This file contains cumulative case counts and deaths from 
# covid-19 for every country in the US. It was current as of
# March 30th.
file = open("covidData_3.30.20.txt", "r")
```

```python
# Find all parishes from Louisiana in this file.
# You should make a list that contains only the parish
# names. This list can be created with a single command.
# Be careful to include ALL the parishes in your search.
parishes = []
for i in file:
    parish = re.search(r",(\w+),Louisiana", i)  # search for a word between two commas and next to the word Louisiana
    if parish:  # if Louisiana is found in the line
        parishes.append(parish.groups()[0])  # add the selected word (parish) to the list
```

```python
# Close the covid data file object
file.close()
```

```python
# Print out the Louisiana search results
for i in parishes:
    print(i)
```

```python
# Create a dictionary that has the Louisiana parish 
# names as keys and the number of times they appear 
# in the data file as values.

# Hint: You can test if a dictionary does or does not
# contain a certain key using:
# myKey in myDict --or-- myKey not in myDict

parish_dict = {}
for i in parishes:  # list from code above
    if i not in parish_dict:  # if there is a "new" parish
        if i != "Unknown":  # remove "unknown" entries from the dictionary
            parish_dict[i] = 1  # add that parish to the dictionary and give it a value of 1
    elif i in parish_dict:  
        parish_dict[i] += 1  # if the parish is already in the dictionary, add 1 to the number of entries
    else: 
        print("ERROR")

for name, number in parish_dict.items():
    print(name + ":", number)
```

```python
# Using your dictionary, figure out which of these parishes has
# the most entries in the file. The one with the most entries
# had the earliest observed covid case.
# Jefferson, Orleans, East Baton Rouge, Plaquemines
comparison = ["Jefferson", "Orleans", "East Baton Rouge", "Plaquemines"]  # list of parishes to compare

max_cases = 0  # number of times a parish has been entered in the list
earliest = ""  # name of that parish
for i in comparison:
    if parish_dict[i] > max_cases:  # if the parish has more entries than the previous maximum, it is now the new max
        max_cases = parish_dict[i]
        earliest = i
print("The first case of covid-19 appeared in", earliest, "Parish.")
    
```

```python
# Figure out how many parishes are in the dictionary. How does 
# this compare to the total number of parishes in the state?
total_parishes = 64  # taken from wikipedia
parishes_affected = len(parish_dict)  # number of values in the dictionary
parishes_not_affected = (total_parishes - parishes_affected)
print("Out of", total_parishes, "parishes in Louisiana, only", parishes_not_affected, 
      "parishes have not been affected by covid-19.")
```

```python
# Now, reopen the data file and use a regex search to find the first instance 
# where a Louisiana parish recorded a covid-19 death. This will be the first 
# line in the file that includes Louisiana and has a death count >0. After 
# your search, print the date and the parish. Be sure to also close the input file.
# Example output: ZZZZ parish reported the first covid-19 death in Louisiana on 2020-ZZ-ZZ

file = open("covidData_3.30.20.txt", "r")

for i in file:
    line = re.search(r"(^.+),(\w+),Louisiana,\d+,\d+,1$", i)  # searches for a line with Louisiana and 1 death
    if line:
        data = line.groups()  # save the values recorded from that line
        break
    
date = data[0]
parish = data[1]

print(parish, "parish reported the first covid-19 death in Louisiana on", date + ".")

file.close()
```

```python
# Now we need to create a new file that only has records from Louisiana. 
# Extract the lines that pertain to Louisiana and write them to a new file.
file = open("covidData_3.30.20.txt", "r")
new_file = open("LA_Cases.txt", "w")

for i in file:
    line = re.search(r"^.+,\w+,Louisiana,\d+,\d+,\d+$", i)
    if line:
        new_file.write(line.group() + "\n")
        
file.close()
new_file.close()
```

```python
# Lastly, let's imagine that we want to do further downstream analyses with
# the Louisiana data, but our downstream analyses require some formatting chanages.
# Read in the lines from your newly created file with only Louisiana records, reformat them, 
# and create a separate file to hold the newly formatted information. Specifically, the 
# dates in the file should be reformatted from looking like this (2020-MM-DD) to looking 
# like this (MM.DD.2020). Also, the fips codes (in the 4th column) should be removed, and
# Louisiana should be abbreviated as LA. So, this file should end up with entries that look
# like this: 03.25.2020,Morehouse,LA,1,0

file = open("LA_Cases.txt", "r")
new_file = open("LA_Cases_New.txt", "w")

for i in file:
    line = re.search(r"(^\d+)-(\d+)-(\d+),(\w+),Louisiana,\d+,(\d+),(\d+$)", i)  # record data from each line containing Louisiana
    if line:
        year = line.groups()[0]  # save values as variables
        month = line.groups()[1]
        day = line.groups()[2]
        parish = line.groups()[3]
        cases = line.groups()[4]
        deaths = line.groups()[5]
        
        new_line = month + "." + day + "." + year + "," + parish + ",LA," + cases + "," + deaths + "\n"
        new_file.write(new_line)
        
file.close()
new_file.close()
```
